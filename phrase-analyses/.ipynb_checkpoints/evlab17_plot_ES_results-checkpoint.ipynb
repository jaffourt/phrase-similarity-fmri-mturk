```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set()
sns.set_palette("Paired")
sns.set_style("white")
```

```python
effect_sum = pd.read_csv("fMRI/SPM_SS_EFFECTS/original/spm_ss_mROI_data.summaries.EffectSize.csv")
effects = pd.unique(effect_sum['Effect'])
effect_inds = pd.read_csv("fMRI/SPM_SS_EFFECTS/originalspm_ss_mROI_data.csv")
#manually reorder effects, if you want
#effects=['I','D','F']
print(effects)
#customize this block for each dataset
nsubjs=len(effect_inds['Subject'].unique())
N = 6
index = np.arange(N)   
title='CompLang Spatial Matching (S>N localizer)'
fig_dims = (60, 30)
color = ["cyan","red","yellow","blue","magenta","white"] 
```

```python
roi_map={'LIFGorb':1,
'LIFG':2,
'LMFG':3,
'LAntTemp':4,
'LPostTemp':5,
'LAngG':6}
effect_sum=effect_sum.replace({"ROI": roi_map})
effect_inds=effect_inds.replace({"ROI": roi_map})
```

```python
SMALL_SIZE = 40
MEDIUM_SIZE = 45
BIGGER_SIZE = 60

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
data_title=title+' n=%d' % nsubjs
```

```python
#parallel lists, assumes that effect_mean_error[0] is associated with effect_mean[0]
#a fair assumption given the 'for' loop
effect_mean = list()
effect_mean_errors = list()
for effect in effects:
    effect_mean.append(effect_sum.loc[(effect_sum.ROI<N+1)&(effect_sum['Effect']==effect)]['MeanEffect'])
    effect_mean_errors.append(effect_sum.loc[(effect_sum.ROI<N+1)&(effect_sum['Effect']==effect)]['StderrEffect'])
eoi_ind = pd.DataFrame().reindex_like(effect_inds)
eoi_ind = eoi_ind.dropna(how='any',axis=0) 
eoi_inds = eoi_ind.append(effect_inds.loc[(effect_inds.ROI<N+1)])
for effect in effects:
    eoi_ind = eoi_ind.append(eoi_inds[eoi_inds['Effect']==effect])
```

```python
fig, ax = plt.subplots(figsize=fig_dims)
capsize = 5
sns.stripplot(x="ROI", y="EffectSize", hue='Effect',data=eoi_ind, 
              alpha=1.0, 
              palette = "Paired", 
              dodge=True, 
              size=5, 
              ax=ax)
space = np.linspace(-len(effects)/2,len(effects)/2,len(effects))
width = (1/float(len(effects)))/2.5
for ind, effect in enumerate(effect_mean):
    p1 = ax.bar(index+width*space[ind], effect, width, edgecolor='black',yerr=effect_mean_errors[ind], capsize=capsize)

# ax.set(xticklabels=['L_TE11','L_TE12','R_TE11','R_TE12'])
ax.set(xticklabels=['LIFGorb','LIFG','LMFG','LAntTemp','LPostTemp','LAngG']) #,'R_IFGorb','R_IFG','R_MFG','R_AntTemp','R_PostTemp','R_AngG'])
ax.set(ylabel='Percent BOLD signal change',
      title=data_title,xlabel='fROIs')
ax.set(ylim=[-5.5, 10.5])
ax.legend(bbox_to_anchor=(.85, .85), ncol=3, fancybox=True, shadow=True, markerscale=3)
```
