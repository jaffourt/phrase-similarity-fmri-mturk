```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
#get the original subj data
original = list()
for file in os.listdir(os.getcwd()):
    if file == 'rankings.csv':
        file_contents = list()
        with open(file,'r') as f:
            for line in f:
                line = line.strip()
                line = line.split(',')
                file_contents.append(line)
for subjID in range(1,26):
    a_ = list()
    for dp in file_contents:
        if dp[3] == str(subjID):
            a_.append(dp[0:3])
    original.append(a_)

#get the 2019 data
subjs_2019 = list()
for file in os.listdir(os.getcwd()):
    if len(file) == 20:
        file_contents = list()
        with open(file,'r') as f:
            for line in f:
                line = line.strip()
                line = line.split(',')
                if not line[2] == 'ranking':
                    file_contents.append(line)
        subjs_2019.append(file_contents)

if len(subjs_2019[0]) == len(original[0]):
    print "Ladies and gentlemen, we got 'em"
```

```python
#pool the 2019 data
avg_2019 = [[0]*4 for _ in range(1,31)]
for subj in subjs_2019:
    for db in subj:
        avg_2019[int(db[0])-1][int(db[1])-2] += int(db[2])
        
#pool the 2019 data
avg_original = [[0]*4 for _ in range(1,31)]
for subj in original:
    for db in subj:
        avg_original[int(db[0])-1][int(db[1])-2] += int(db[2])
```

```python
# calculate the correlation
# BE CAREFUL with logic here, it's easy to get confused and do this incorrectly:
# you measure the correlation of n=1 for each set (120) 
# by comparing it to the averages of n=49 for each set (120)
R_values = list()
for subj in subjs_2019:  
    #subj is 120 rankings (4 per 30 base)
    #subj[0:119][0] are the bases
    #subj[0:119][1] are the transformations
    #subj[0:119][2] are the rankings
    #copy avg_2019, recalculate the averages without subj
    copy_of_avg_2019 = [[0]*4 for _ in range(1,31)]
    copy_of_subjs_2019 = list(subjs_2019)
    copy_of_subjs_2019.remove(subj)
    #re do all the calculations like before
    M_total = list()
    P_total = list()
    N_total = list()
    A_total = list()
    # calculate the ranking
    for copy_subj in copy_of_subjs_2019:
        for q in copy_subj:  
            if q[0].isdigit():
                q = map(int, q)
                if q[1] == 2:
                    M_total.append(q[2])
                elif q[1] == 3:
                    N_total.append(q[2])
                elif q[1] == 4:
                    P_total.append(q[2])
                elif q[1] == 5:
                    A_total.append(q[2])
    #calculate averages
    #and standard error of the mean
    avg = [0,0,0,0]
    avg[0] = np.mean(M_total)
    avg[1] = np.mean(N_total)
    avg[2] = np.mean(P_total)
    avg[3] = np.mean(A_total)
    #Done! Now we have the averages of the population w/o the current sample "subj"
    sum = 0
    mag = 0
    for i in range(1,31):
        prod_sum = 1
        prod_mag = 1
        for data in subj:
            if int(data[0]) == i:
                prod_sum *= int(data[2])-avg[int(data[1])-2]
                prod_mag *= np.abs(int(data[2])-avg[int(data[1])-2])
        sum+=prod_sum
        mag+=prod_mag
    R_values.append(sum/mag)
```
